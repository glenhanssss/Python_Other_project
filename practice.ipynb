```python
#GAME TEBAK ANGKA 1-100 dikasih tau besar dan kecil, 5 kali kesempatan
import random

print("Selamat datang di game tebak angka!")
print("Saya telah memilih angka secara acak dari 1 hingga 100.")
print("Anda memiliki 5 kesempatan untuk menebak angka tersebut.")
print("Selamat bermain!")

angka_rahasia = random.randint(1, 100)
kesempatan = 5
tebakan = None

while kesempatan > 0 and tebakan != angka_rahasia:
    print("Kesempatan anda:", kesempatan)
    tebakan = int(input("Silakan masukkan angka tebakan anda: "))
    if tebakan < angka_rahasia:
        print("Angka tebakan anda terlalu kecil!")
    elif tebakan > angka_rahasia:
        print("Angka tebakan anda terlalu besar!")
    else:
        print("Selamat! Angka tebakan anda benar!")
    kesempatan -= 1

if kesempatan == 0:
    print("Maaf, kesempatan anda telah habis.")
    print("Angka rahasia yang saya pilih adalah:", angka_rahasia)
    
print("Terima kasih sudah bermain!")
```

```python
#GAME TEBAK ANGKA 1-10 ga dikasih tau besar dan kecil, 3 kali kesempatan
import random

# Memilih angka acak
angka = random.randint(1, 10)

# Memberikan tiga kesempatan
kesempatan = 3

# Looping untuk mencoba menebak angka
while kesempatan > 0:
    # Meminta input dari user
    tebakan = int(input("Tebak angka (1-10): "))
    
    # Mengecek apakah tebakan benar atau salah
    if tebakan == angka:
        print("Anda benar!")
        break
    else:
        print("Anda salah!")
        kesempatan -= 1

# Jika kesempatan habis
if kesempatan == 0:
    print("Anda kalah. Angka yang benar adalah", angka)
```

```python
import random

choices = ['rock', 'paper', 'scissors']

while True:
    # Get user input
    user_choice = input("Choose rock, paper, or scissors (or q to quit): ").lower()

    # Quit if user enters 'q'
    if user_choice == 'q':
        break

    # Check if user input is valid
    if user_choice not in choices:
        print("Invalid input. Please choose rock, paper, or scissors.")
        continue

    # Generate computer's random choice
    computer_choice = random.choice(choices)

    # Determine the winner
    if user_choice == computer_choice:
        print("Tie!")
    elif user_choice == 'rock' and computer_choice == 'scissors':
        print("You win! Rock smashes scissors.")
    elif user_choice == 'paper' and computer_choice == 'rock':
        print("You win! Paper covers rock.")
    elif user_choice == 'scissors' and computer_choice == 'paper':
        print("You win! Scissors cut paper.")
    else:
        print("You lose. Better luck next time!")

print("Thanks for playing!")
```

```python
#Game Ngitung Mudah, Sedang, Sulit
import random

print("Selamat datang di game Matematika")
print("Silahkan pilih level kesulitan:")
print("1 - Mudah")
print("2 - Sedang")
print("3 - Sulit")

level = int(input("Masukkan pilihan level: "))

if level == 1:
    min_range = 1
    max_range = 10
    num_of_questions = 5
elif level == 2:
    min_range = 1
    max_range = 20
    num_of_questions = 10
elif level == 3:
    min_range = 1
    max_range = 50
    num_of_questions = 15
else:
    print("Pilihan level tidak valid.")
    exit()

score = 0

for i in range(num_of_questions):
    num1 = random.randint(min_range, max_range)
    num2 = random.randint(min_range, max_range)
    operator = random.choice(["+", "-", "*"])

    if operator == "+":
        correct_answer = num1 + num2
    elif operator == "-":
        correct_answer = num1 - num2
    else:
        correct_answer = num1 * num2

    print("Pertanyaan ke-", i+1)
    print("Berapa hasil dari {} {} {}?".format(num1, operator, num2))
    answer = int(input("Jawaban Anda: "))

    if answer == correct_answer:
        print("Jawaban Anda benar!")
        score += 1
    else:
        print("Jawaban Anda salah. Jawaban yang benar adalah", correct_answer)

print("Permainan selesai!")
print("Skor Anda adalah", score, "dari", num_of_questions)
```
